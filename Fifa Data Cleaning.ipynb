```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print (row['Contract'])
```

```python
def extract_contract(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date= np.nan
        end_date= np.nan
        contract_length=0
    else:
       start_date, end_date = contract.split(' ~ ')
       start_year = int(start_date[:4])
       end_year = int(end_date[:4])  
       contract_length = end_year - start_year
    return start_date, end_date, contract_length


#Applying function to contract columns, creating three columns

new_cols = ['Contract Start' , 'Contract End' , 'Contract Length(years)']
new_data= fifa['Contract'].apply(lambda x: pd.Series(extract_contract(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i , column=new_cols[i] , value=new_data[i])
```

```python
#contract categories

def categories_contract(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

#Contract status column

fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract status' , fifa['Contract'].apply(categories_contract))
fifa.sample(3)
```

```python
#cleaning Height column
def convert_height(height):
    if "cm" in height:
     return int(height.strip("cm"))
    else:
     feet,inches = height.split("'")
     total_inches= int(feet)*12+ int(inches.strip('"'))
     return round(total_inches*2.54)
#Applying function to height
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns={'Height' : "Height(cm)"})
fifa.sample(1)
```

```python
#Cleaning weight columns

def convert_weights(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.205)

#Applying function to weight columns

fifa['Weight'] = fifa['Weight'].apply(convert_weights)
fifa['Weight'].unique()
```
